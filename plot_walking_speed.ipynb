```python
wk_weights = []
with open("output/results.txt", encoding="utf-8") as f:
    # starts_list = [x.rstrip() for x in f]
    line_count = 1
    for line in f:
        if line_count % 4 == 0:
            weight = float(line.split("=")[1].strip())
            wk_weights.append(weight)
        line_count += 1

print(wk_weights)
```

```python
import matplotlib.pyplot as plt
import numpy as np
ax = plt.gca()
xpoints = range(1, len(wk_weights)+1)
ypoints =  wk_weights
ax.bar(xpoints, ypoints, width=0.3, color='k')
plt.plot(np.array([0, 40]), np.array([1, 1]), 'k--', lw=1)
plt.plot(np.array([0, 40]), np.array([100, 100]), 'k--', lw=1)
for i in range(len(ypoints)):
    if ypoints[i] > 1 and ypoints[i] < 100:
        plt.scatter(xpoints[i], ypoints[i], c='b', s=10, marker='x')
    elif ypoints[i] == 1 or ypoints[i] == 100:
        plt.scatter(xpoints[i], ypoints[i], c='r', s=10, marker='x')
    # plt.plot(np.array([i+1, i+1]), np.array([1, wk_weights[i]]), 'k--', lw=1)
# plt.scatter(xpoints, ypoints, c='r', s=10, marker='x')
ax.set_ylim([0.8, 5])
plt.show()
# plt.savefig('output/weights_lim.png', dpi=300, bbox_inches='tight')
```

```python
age_groups = range(20, 81, 10)
female_speed = [1.50, 1.49, 1.47, 1.43, 1.42, 1.30, 1.18]
male_speed = [1.56, 1.55, 1.48, 1.46, 1.46, 1.36, 1.25]

plt.plot(age_groups, female_speed, 'r', lw=1, label="Female")
plt.plot(age_groups, male_speed, 'b', lw=1, label="Male")
plt.plot(age_groups, [1.34]*7, 'k--', lw=1)

plt.annotate('1.34 m/s = 3 mile per hour', xy=(20, 1.34), xytext=(20, 1.35), color='k', fontsize=8)
plt.xlabel('Age')
plt.ylabel('Walking speed (m/s)')
plt.legend(loc="upper right")
plt.savefig('output/age_trend.png', dpi=300, bbox_inches='tight')
```
