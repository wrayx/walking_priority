```python
import json

london_tube_lines = (
	'Bakerloo',
	'Central',
	'District',
	'Circle',
    'DLR',
    'Hammersmith and City'
	'Jubilee',
	'Metropolitan',
	'Northern',
    'Overground',
	'Piccadilly',
	'Victoria',
	'Waterloo and City',
    'Elizabeth'
)
  
# Opening JSON file
f = open('input/LondonTubeLines.json')

data_minimal = {"features":[]}
  
# returns JSON object as 
# a dictionary
data = json.load(f)
print(len(data['features']))

# Iterating through the json list
for tube_route in data['features']:
    if tube_route['properties']['name'].startswith(london_tube_lines):
        print(tube_route['properties']['name'])
        data_minimal['features'].append(tube_route)

print(len(data_minimal['features']))
with open('input/LondonTubeLines_minimal.json', 'w', encoding='utf-8') as f:
    json.dump(data_minimal, f, ensure_ascii=False, indent=4)

# Closing file
f.close()
```
